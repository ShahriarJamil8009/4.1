```python
## 1st step: import necessary libreries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
## 2nd step: assign and resize image
originalImage = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
## 3rd step: show image in ploats with brightness enhancement
plt.figure(figsize=(12, 6))

# Original Image Visualization
plt.subplot(1, 2, 1)
plt.title('original Imag')
plt.imshow(originalImage, cmap='gray')

# enhance Image Visualization
plt.subplot(1, 2, 2)
plt.title('Enhenced Image')
enhencedImage = np.where((originalImage >= 10) & (originalImage <= 50), np.clip(originalImage + 40, 0, 255), originalImage)
plt.imshow(enhencedImage, cmap='gray')
```

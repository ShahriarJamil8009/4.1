```python
## 1st step: import necessary libreries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
## 2nd step: assign and resize image
originalImage = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
## 3rd step: show image in ploats
plt.figure(figsize=(12, 6))

# Original Image Visualization
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(originalImage, cmap='gray')

# MSB Image Visualization
plt.subplot(1, 3, 2)
plt.title("MSB Image")
msbImage = originalImage & 0b11100000
plt.imshow(msbImage, cmap='gray')

# Differecne Image Visualization
plt.subplot(1, 3, 3)
plt.title("Difference Image")
differenceImage = abs(originalImage - msbImage)
plt.imshow(differenceImage, cmap='gray')
```

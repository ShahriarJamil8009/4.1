```python
## 1st step: import necessary libreries
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
## 2nd step: assign and resize image
originalImage = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
originalImage = cv2.resize(originalImage, (512, 512))
```

```python
## 4th step: create histograme
def generateHistogram(image):
    histogram = np.zeros(256)   # Creates an array of 256 zeros to store the frequency of each pixel intensity value
    height, width = image.shape

    for i in range(height):
        for j in range(width):
            histogram[image[i, j]] += 1
    return histogram
```

```python
## 3rd step: show image in ploats
plt.figure(figsize=(16, 8))

# Original Image Visualization & Histogram
plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(originalImage, cmap='gray')

plt.subplot(2, 2, 2)
plt.title("Histogram (Original Image)")
plt.bar(range(256), generateHistogram(originalImage), width=1.0)

# Thresholding, 
thresholdValue = 128
sigmentedImage = np.where(originalImage > thresholdValue, 255, 0)
sigmentedImage = np.uint8(sigmentedImage)

# Segment Image Visualization
plt.subplot(2, 2, 3)
plt.title("Sigmented Image")
plt.imshow(sigmentedImage, cmap='gray')

plt.subplot(2, 2, 4)
plt.title("Histogram (Sigmented Image)")
plt.bar(range(256), generateHistogram(sigmentedImage), width=1.0)
```
